```python
import numpy as np
import pandas as pd
import math

# -----------------------------
# FIXED ROLL NUMBER
# -----------------------------
roll = 102303736

# Compute a_r and b_r
a_r = 0.05 * (roll % 7)        # = 0.15
b_r = 0.3 * ((roll % 5) + 1)   # = 0.6

# -----------------------------
# LOAD DATA
# -----------------------------
# Change path if needed
df = pd.read_csv("data 2.csv", on_bad_lines='skip', encoding='latin1')

# Use NO2 column
x = pd.to_numeric(df["no2"], errors="coerce").dropna().to_numpy()

# -----------------------------
# TRANSFORMATION
# -----------------------------
z = x + a_r * np.sin(b_r * x)

# -----------------------------
# PARAMETER ESTIMATION (MLE)
# -----------------------------
mu = np.mean(z)
sigma2 = np.mean((z - mu) ** 2)     # population variance (MLE)

lam = 1 / (2 * sigma2)
c = 1 / math.sqrt(2 * math.pi * sigma2)

# -----------------------------
# OUTPUT
# -----------------------------
print("Roll number:", roll)
print("a_r =", a_r)
print("b_r =", b_r)
print("mu =", mu)
print("lambda =", lam)
print("c =", c)

# Consistency check
print("Check c = sqrt(lambda/pi):", math.sqrt(lam / math.pi))
```
